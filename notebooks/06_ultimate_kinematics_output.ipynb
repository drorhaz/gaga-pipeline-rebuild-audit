## Helpers: Quaternion unrolling and renormalization

```python
def unroll_quat(q):
    """Ensure temporal continuity (shortest path); q shape (T, 4)."""
    q = np.asarray(q, dtype=float)
    if q.ndim == 1:
        q = q.reshape(1, -1)
    out = q.copy()
    for i in range(1, len(out)):
        if np.dot(out[i], out[i - 1]) < 0:
            out[i] *= -1
    return out

def renormalize_quat(q):
    """Unit norm per row; q shape (T, 4)."""
    n = np.linalg.norm(q, axis=1, keepdims=True)
    n = np.where(n < 1e-12, 1.0, n)
    return q / n
```

## Root-relative positions and dual-track quaternion processing

For each joint: hierarchical q_rel = inv(parent)*child → unroll → SavGol smooth → renormalize (Track A). Track B: q_zeroed = inv(q_ref_rel)*q_raw_smooth. Then ω from quat_log, α from SavGol derivative on ω. Linear v,a from root-relative positions.

```python
# Root-relative positions
root_pos_cols = [f"{ROOT_SEGMENT}__px", f"{ROOT_SEGMENT}__py", f"{ROOT_SEGMENT}__pz"]
root_pos = df_in[root_pos_cols].values if all(c in df_in.columns for c in root_pos_cols) else None

# Ticket 007b + MUB-NB06: capture which position axis columns are dropped by the
# notna().all() gate. This gate silently removes lin_rel/lin_vel/lin_acc columns
# for any joint with even 1 NaN frame in a given axis. We log the dropped axes
# to validation_report.json (consumed in Cell 20) so downstream feature reliability
# (Ticket 014b) can flag affected sessions. The gate itself is NOT changed
# (per MUB Option A; gate upgrade deferred to post-Minimal-v1).
_t007b_all_pos_axis_cols = [c for c in df_in.columns if c.endswith(('__px', '__py', '__pz'))]
pos_cols = [c for c in _t007b_all_pos_axis_cols if df_in[c].notna().all()]
_t007b_dropped_axes = {}
for _c_t007b in _t007b_all_pos_axis_cols:
    if _c_t007b not in pos_cols:
        _joint_t007b = _c_t007b.split('__')[0]
        _axis_t007b = _c_t007b[-1]  # 'x', 'y', or 'z'
        _nan_count_t007b = int(df_in[_c_t007b].isna().sum())
        _n_total_t007b = int(len(df_in))
        _t007b_dropped_axes.setdefault(_joint_t007b, {})[_axis_t007b] = {
            'nan_count': _nan_count_t007b,
            'nan_fraction': float(_nan_count_t007b) / float(_n_total_t007b) if _n_total_t007b > 0 else 0.0,
        }
_t007b_lin_kine_diagnostics = {
    'gate': "notna().all() per axis column (NB06 Cell 7)",
    'note': 'Any NaN in a position axis column silently drops 5 derived columns for that joint+axis. See MUB_NB06_lin_kine_nan_gate.',
    'dropped_axes': _t007b_dropped_axes,
    'n_joints_with_dropped_axes': len(_t007b_dropped_axes),
    'total_axes_silently_dropped': sum(len(v) for v in _t007b_dropped_axes.values()),
}
print(f"[Ticket 007b/MUB] lin_kine NaN gate: {len(_t007b_dropped_axes)} joints with dropped axes "
      f"({sum(len(v) for v in _t007b_dropped_axes.values())} total axes)")
axis_idx = {'__px': 0, '__py': 1, '__pz': 2}
pos_rel = {}
if root_pos is not None:
    for col in pos_cols:
        seg, suffix = col.split('__')[0], '__' + col.split('__')[1]
        idx = axis_idx.get(suffix, 0)
        pos_rel[col] = df_in[col].values - root_pos[:, idx]

# Master result: time_s first
result = {'time_s': df_in['time_s'].values}
validation_rows = []
OMEGA_THRESH = 3000.0
LIN_ACC_THRESH = 20000.0
```

## Master Parquet Feature Set (Gold Standard for ML/HMM/RQA)

The exported `kinematics_master.parquet` contains the following feature categories per joint/segment:

### A. Orientation (Posture) - Per Joint
- `{joint}__raw_rel_qx, qy, qz, qw`: Original hierarchical quaternions (parent→child)
- `{joint}__zeroed_rel_qx, qy, qz, qw`: T-pose normalized quaternions
- `{joint}__zeroed_rel_rotvec_x, y, z`: **Rotation Vector in rad** (crucial for HMM/ML)
- `{joint}__zeroed_rel_rotmag`: **Geodesic distance from T-pose in deg** (crucial for RQA)

### B. Angular Kinematics (Zeroed Track) - Per Joint
- `{joint}__zeroed_rel_omega_x, y, z`: Angular velocity vector in deg/s
- `{joint}__zeroed_rel_omega_mag`: **Angular velocity magnitude in deg/s** (invariant)
- `{joint}__zeroed_rel_alpha_x, y, z`: Angular acceleration vector in deg/s²
- `{joint}__zeroed_rel_alpha_mag`: **Angular acceleration magnitude in deg/s²**

### C. Linear Kinematics (Root-Relative) - Per Segment
- `{segment}__lin_rel_px, py, pz`: Root-relative position in mm
- `{segment}__lin_vel_rel_x, y, z`: Linear velocity vector in mm/s
- `{segment}__lin_vel_rel_mag`: **Linear velocity magnitude in mm/s**
- `{segment}__lin_acc_rel_x, y, z`: Linear acceleration vector in mm/s²
- `{segment}__lin_acc_rel_mag`: **Linear acceleration magnitude in mm/s²**

### D. Whole-Body Center of Mass
- `wbc_com_x, wbc_com_y, wbc_com_z`: Whole-body CoM position (root-relative, mm)
- Computed using de Leva (1996) segment mass fractions with missing-segment compensation

### E. Parquet Metadata (PyArrow embedded)
- `run_id`, `subject_id`, `subject_height_cm`, `subject_mass_kg`, `pipeline_version`, `processing_timestamp`

**Note:** Raw angular features (raw_rel_omega, raw_rel_alpha) are also included for comparison and validation.

```python
for joint_name, info in kinematics_map.items():
    parent_name = info.get('parent')
    q_c_cols = [f"{joint_name}__qx", f"{joint_name}__qy", f"{joint_name}__qz", f"{joint_name}__qw"]
    if not all(c in df_in.columns for c in q_c_cols):
        continue
    q_c = df_in[q_c_cols].values
    if parent_name is not None:
        q_p_cols = [f"{parent_name}__qx", f"{parent_name}__qy", f"{parent_name}__qz", f"{parent_name}__qw"]
        if not all(c in df_in.columns for c in q_p_cols):
            continue
        q_p = df_in[q_p_cols].values
        rot_p = R.from_quat(q_p)
        rot_c = R.from_quat(q_c)
        q_rel = (rot_p.inv() * rot_c).as_quat()
    else:
        q_rel = q_c.copy()
    q_rel = unroll_quat(q_rel)
    _jw = W_LEN_MAP.get(joint_name, W_LEN)
    _jw_src = "adaptive" if joint_name in W_LEN_MAP else "global"
    for ax in range(4):
        q_rel[:, ax] = chunked_savgol(q_rel[:, ax], _jw, SG_POLYORDER, deriv=0)
    q_raw_smooth = renormalize_quat(q_rel)
    # Manifold guard: any non-unit or non-finite rows become NaN (safe for downstream)
    _norms = np.linalg.norm(q_raw_smooth, axis=1)
    _bad = ~np.isfinite(_norms) | (_norms < 0.99)
    if _bad.any():
        q_raw_smooth[_bad] = np.nan
    q_ref_c = np.array([ref_pose[f"{joint_name}__qx"], ref_pose[f"{joint_name}__qy"], ref_pose[f"{joint_name}__qz"], ref_pose[f"{joint_name}__qw"]])
    if parent_name is not None:
        q_ref_p = np.array([ref_pose[f"{parent_name}__qx"], ref_pose[f"{parent_name}__qy"], ref_pose[f"{parent_name}__qz"], ref_pose[f"{parent_name}__qw"]])
        q_rel_ref = (R.from_quat(q_ref_p).inv() * R.from_quat(q_ref_c)).as_quat()
    else:
        q_rel_ref = q_ref_c
    rot_ref = R.from_quat(q_rel_ref)
    q_zeroed = (rot_ref.inv() * R.from_quat(q_raw_smooth)).as_quat()
    q_zeroed = renormalize_quat(q_zeroed)
    omega_raw_rad, alpha_raw_rad = compute_omega_and_alpha(q_raw_smooth, FS, method=OMEGA_METHOD, frame=OMEGA_FRAME, savgol_window=_jw, savgol_poly=SG_POLYORDER)
    omega_zeroed_rad, alpha_zeroed_rad = compute_omega_and_alpha(q_zeroed, FS, method=OMEGA_METHOD, frame=OMEGA_FRAME, savgol_window=_jw, savgol_poly=SG_POLYORDER)
    omega_raw_deg = np.degrees(omega_raw_rad)
    omega_zeroed_deg = np.degrees(omega_zeroed_rad)
    alpha_raw = np.degrees(alpha_raw_rad)
    alpha_zeroed = np.degrees(alpha_zeroed_rad)
    for ax, letter in enumerate(['x', 'y', 'z']):
        result[f"{joint_name}__raw_rel_omega_{letter}"] = omega_raw_deg[:, ax]
        result[f"{joint_name}__zeroed_rel_omega_{letter}"] = omega_zeroed_deg[:, ax]
    for ax, letter in enumerate(['x', 'y', 'z']):
        result[f"{joint_name}__raw_rel_alpha_{letter}"] = alpha_raw[:, ax]
        result[f"{joint_name}__zeroed_rel_alpha_{letter}"] = alpha_zeroed[:, ax]
    for ax, letter in enumerate(['x', 'y', 'z', 'w']):
        result[f"{joint_name}__raw_rel_q{letter}"] = q_raw_smooth[:, ax]
        result[f"{joint_name}__zeroed_rel_q{letter}"] = q_zeroed[:, ax]
    mag_omega_raw = np.linalg.norm(omega_raw_deg, axis=1)
    mag_omega_zeroed = np.linalg.norm(omega_zeroed_deg, axis=1)
    geodesic_deg = np.degrees(2 * np.arccos(np.clip(np.abs((q_raw_smooth * q_zeroed).sum(axis=1)), 0, 1)))
    vel_align = 100.0 * pearsonr(mag_omega_raw, mag_omega_zeroed)[0] if np.std(mag_omega_raw) > 1e-10 else 100.0
    
    # Compute magnitude of angular acceleration
    mag_alpha_raw = np.linalg.norm(alpha_raw, axis=1)
    
    validation_rows.append({
        'joint': joint_name,
        'sg_w_len': _jw,
        'sg_w_source': _jw_src,
        'geodesic_offset_std': round(float(np.std(geodesic_deg)), 6),
        'velocity_alignment_pct': round(vel_align, 2),
        'max_omega_deg_s': round(float(np.max(mag_omega_raw)), 2),
        'mean_omega_deg_s': round(float(np.mean(mag_omega_raw)), 2),
        'median_omega_deg_s': round(float(np.median(mag_omega_raw)), 2),
        'exceeded_omega_threshold': bool(np.max(mag_omega_raw) > OMEGA_THRESH),
        'max_alpha_deg_s2': round(float(np.max(mag_alpha_raw)), 2),
        'mean_alpha_deg_s2': round(float(np.mean(mag_alpha_raw)), 2),
    })
_w_summary = {r['joint']: r['sg_w_len'] for r in validation_rows}
print(f"Processed {len(validation_rows)} joints (dual-track omega, alpha, quaternions).")
print(f"  Per-joint SavGol W: {_w_summary}")
```

```python
# Linear velocity and acceleration from root-relative positions (adaptive SavGol)
for col in pos_cols:
    seg, suffix = col.split('__')[0], '__' + col.split('__')[1]
    if col not in pos_rel:
        continue
    pr = pos_rel[col]
    _sw = W_LEN_MAP.get(seg, W_LEN)
    vel = chunked_savgol(pr, _sw, SG_POLYORDER, deriv=1, delta=dt)
    acc = chunked_savgol(pr, _sw, SG_POLYORDER, deriv=2, delta=dt)
    axis_letter = col[-1]
    result[f"{seg}__lin_vel_rel_{axis_letter}"] = vel
    result[f"{seg}__lin_acc_rel_{axis_letter}"] = acc
segments_with_pos = set(col.split('__')[0] for col in pos_cols if col in pos_rel)
lin_audit = []
for seg in segments_with_pos:
    # Linear acceleration magnitude
    cx, cy, cz = f"{seg}__lin_acc_rel_x", f"{seg}__lin_acc_rel_y", f"{seg}__lin_acc_rel_z"
    if cx in result and cy in result and cz in result:
        mag_acc = np.linalg.norm(np.column_stack([result[cx], result[cy], result[cz]]), axis=1)
        max_acc = round(float(np.max(mag_acc)), 2)
    else:
        max_acc = 0.0
    
    # Linear velocity magnitude
    vx, vy, vz = f"{seg}__lin_vel_rel_x", f"{seg}__lin_vel_rel_y", f"{seg}__lin_vel_rel_z"
    if vx in result and vy in result and vz in result:
        mag_vel = np.linalg.norm(np.column_stack([result[vx], result[vy], result[vz]]), axis=1)
        max_vel = round(float(np.max(mag_vel)), 2)
    else:
        max_vel = 0.0
    
    _seg_w = W_LEN_MAP.get(seg, W_LEN)
    lin_audit.append({
        'segment': seg,
        'sg_w_len': _seg_w,
        'sg_w_source': "adaptive" if seg in W_LEN_MAP else "global",
        'max_lin_vel_mm_s': max_vel,
        'max_lin_acc_mm_s2': max_acc,
        'exceeded_lin_acc_threshold': bool(max_acc > LIN_ACC_THRESH)
    })
_lin_w_summary = {a['segment']: a['sg_w_len'] for a in lin_audit}
print(f"Linear v,a computed for {len(pos_cols)} position components; {len(lin_audit)} segments audited.")
print(f"  Per-segment SavGol W: {_lin_w_summary}")
```

```python
# Add magnitude and rotation vector features to result (required for ML/HMM/RQA)
# These are the master features as per specification

# A. Add rotation vector (rotvec) and rotation magnitude (rotmag) for zeroed quaternions
for joint_name in kinematics_map:
    qc = [f"{joint_name}__zeroed_rel_qx", f"{joint_name}__zeroed_rel_qy", f"{joint_name}__zeroed_rel_qz", f"{joint_name}__zeroed_rel_qw"]
    if not all(c in result for c in qc):
        continue
    q_zeroed = np.column_stack([result[c] for c in qc])
    # Rotation vector (rotvec) in radians
    rotvec = R.from_quat(q_zeroed).as_rotvec()
    for ax, letter in enumerate(['x', 'y', 'z']):
        result[f"{joint_name}__zeroed_rel_rotvec_{letter}"] = rotvec[:, ax]
    # Rotation magnitude (geodesic distance from T-pose) in degrees
    result[f"{joint_name}__zeroed_rel_rotmag"] = np.degrees(np.linalg.norm(rotvec, axis=1))

# B. Add angular velocity magnitude (omega_mag) and angular acceleration magnitude (alpha_mag)
for joint_name in kinematics_map:
    # Zeroed omega magnitude
    omega_cols = [f"{joint_name}__zeroed_rel_omega_x", f"{joint_name}__zeroed_rel_omega_y", f"{joint_name}__zeroed_rel_omega_z"]
    if all(c in result for c in omega_cols):
        omega_vec = np.column_stack([result[c] for c in omega_cols])
        result[f"{joint_name}__zeroed_rel_omega_mag"] = np.linalg.norm(omega_vec, axis=1)
    
    # Zeroed alpha magnitude
    alpha_cols = [f"{joint_name}__zeroed_rel_alpha_x", f"{joint_name}__zeroed_rel_alpha_y", f"{joint_name}__zeroed_rel_alpha_z"]
    if all(c in result for c in alpha_cols):
        alpha_vec = np.column_stack([result[c] for c in alpha_cols])
        result[f"{joint_name}__zeroed_rel_alpha_mag"] = np.linalg.norm(alpha_vec, axis=1)

# C. Add linear velocity and acceleration magnitudes
segments_with_pos = set(col.split('__')[0] for col in pos_cols if col in pos_rel)
for seg in segments_with_pos:
    # Linear velocity magnitude
    vel_cols = [f"{seg}__lin_vel_rel_x", f"{seg}__lin_vel_rel_y", f"{seg}__lin_vel_rel_z"]
    if all(c in result for c in vel_cols):
        vel_vec = np.column_stack([result[c] for c in vel_cols])
        result[f"{seg}__lin_vel_rel_mag"] = np.linalg.norm(vel_vec, axis=1)
    
    # Linear acceleration magnitude
    acc_cols = [f"{seg}__lin_acc_rel_x", f"{seg}__lin_acc_rel_y", f"{seg}__lin_acc_rel_z"]
    if all(c in result for c in acc_cols):
        acc_vec = np.column_stack([result[c] for c in acc_cols])
        result[f"{seg}__lin_acc_rel_mag"] = np.linalg.norm(acc_vec, axis=1)

# D. Add root-relative positions (lin_rel_px, py, pz)
for col in pos_cols:
    seg, suffix = col.split('__')[0], '__' + col.split('__')[1]
    if col in pos_rel:
        axis_letter = col[-1]
        result[f"{seg}__lin_rel_p{axis_letter}"] = pos_rel[col]

print(f"Added derived features: rotvec (x,y,z), rotmag, omega_mag, alpha_mag, vel_mag, acc_mag, and lin_rel positions for {len(validation_rows)} joints and {len(segments_with_pos)} segments.")
```

```python
# ============================================================
# F. PER-FRAME ARTIFACT BOOLEAN FLAGS (Dual-Criteria Detection)
# ============================================================
# Combines Angular and Linear velocity thresholds for robust artifact detection
# Angular: Catches gimbal issues and non-physical rotations
# Linear: "Teleportation Detector" for marker swaps and tracking loss

print("\n" + "="*80)
print("ARTIFACT DETECTION (Dual-Criteria: Angular + Linear)")
print("="*80)

# ISB-Based Thresholds
THRESH_ARTIFACT = {
    'rotation_mag_deg': 140.0,          # Extreme joint angle (near anatomical limits)
    'angular_velocity_deg_s': 800.0,    # High angular velocity (gimbal/sensor issues)
    'linear_velocity_mm_s': 3000.0,     # Ballistic limit (teleportation detector)
}

artifact_count = 0
segments_flagged = 0

# Joint-level artifacts (rotation + angular velocity)
for joint_name in kinematics_map:
    artifact_mask = np.zeros(len(result['time_s']), dtype=bool)
    
    # Criterion 1: Extreme rotation magnitude
    rotmag_col = f"{joint_name}__zeroed_rel_rotmag"
    if rotmag_col in result:
        artifact_mask |= (result[rotmag_col] >= THRESH_ARTIFACT['rotation_mag_deg'])
    
    # Criterion 2: High angular velocity
    omega_mag_col = f"{joint_name}__zeroed_rel_omega_mag"
    if omega_mag_col in result:
        artifact_mask |= (result[omega_mag_col] >= THRESH_ARTIFACT['angular_velocity_deg_s'])
    
    # Store flag
    result[f"{joint_name}__is_artifact"] = artifact_mask
    artifact_count += np.sum(artifact_mask)
    if np.any(artifact_mask):
        segments_flagged += 1

# Segment-level artifacts (linear velocity - teleportation detector)
segments_with_pos = set(col.split('__')[0] for col in result.keys() if '__lin_rel_px' in col)
for seg in segments_with_pos:
    # Initialize with joint flags if segment = joint
    if seg in kinematics_map and f"{seg}__is_artifact" in result:
        artifact_mask = result[f"{seg}__is_artifact"].copy()
    else:
        artifact_mask = np.zeros(len(result['time_s']), dtype=bool)
    
    # Criterion 3: Ballistic linear velocity (marker swap/loss)
    linvel_mag_col = f"{seg}__lin_vel_rel_mag"
    if linvel_mag_col in result:
        artifact_mask |= (result[linvel_mag_col] >= THRESH_ARTIFACT['linear_velocity_mm_s'])
    
    # Update or create flag
    result[f"{seg}__is_artifact"] = artifact_mask
    artifact_count += np.sum(artifact_mask)

print(f"\n[OK] Artifact flags added for {len([k for k in result.keys() if '__is_artifact' in k])} segments/joints")
print(f"     Total artifact frames: {artifact_count:,}")
print(f"     Segments with artifacts: {segments_flagged}")
print(f"     Detection criteria:")
print(f"       - Rotation magnitude > {THRESH_ARTIFACT['rotation_mag_deg']:.0f} deg")
print(f"       - Angular velocity > {THRESH_ARTIFACT['angular_velocity_deg_s']:.0f} deg/s")
print(f"       - Linear velocity > {THRESH_ARTIFACT['linear_velocity_mm_s']:.0f} mm/s (teleportation)")

# Ticket 011: Load per-frame Hampel OR mask from S04 sidecar and propagate to
# {joint}__is_hampel_outlier column (Option B: single OR boolean across all 19 joints).
# The mask is True at frame t if ANY position column was Hampel-replaced at frame t in S04.
# Graceful fallback to all-False if the sidecar is missing (Phase 11.5 R3).
_t011_mask_path = Path(DERIV_FILTERED) / f"{RUN_ID}__s04_hampel_or_mask.npy"
_t011_n_frames = len(result['time_s'])
try:
    _t011_mask_raw = np.load(str(_t011_mask_path))
    if len(_t011_mask_raw) == _t011_n_frames:
        _t011_hampel_or = _t011_mask_raw.astype(bool)
    else:
        import logging as _lg_t011
        _lg_t011.getLogger('ticket011').warning(
            "[Ticket 011] Hampel mask length mismatch: mask=%d vs n_frames=%d for %s; falling back to all-False",
            len(_t011_mask_raw), _t011_n_frames, RUN_ID
        )
        _t011_hampel_or = np.zeros(_t011_n_frames, dtype=bool)
    _n_t011_flagged = int(np.sum(_t011_hampel_or))
    print(f"[Ticket 011] Hampel OR mask loaded: {_n_t011_flagged}/{_t011_n_frames} frames flagged "
          f"({100.0*_n_t011_flagged/max(_t011_n_frames,1):.3f}%)")
except Exception as _e_t011:
    import logging as _lg_t011b
    _lg_t011b.getLogger('ticket011').warning(
        "[Ticket 011] Hampel mask not found for %s (%s); falling back to all-False", RUN_ID, _e_t011
    )
    _t011_hampel_or = np.zeros(_t011_n_frames, dtype=bool)
    print(f"[Ticket 011] WARNING: Hampel mask not found for {RUN_ID}; is_hampel_outlier set to all-False")

# Apply OR mask to ALL joint and segment columns
for joint_name in kinematics_map:
    result[f"{joint_name}__is_hampel_outlier"] = _t011_hampel_or.copy()
for seg in segments_with_pos:
    if f"{seg}__is_hampel_outlier" not in result:
        result[f"{seg}__is_hampel_outlier"] = _t011_hampel_or.copy()

print(f"[OK] Hampel OR mask applied to {len([k for k in result if '__is_hampel_outlier' in k])} columns")
print("="*80)
```

```python
import pyarrow as pa
import pyarrow.parquet as pq
from datetime import datetime

df_master = pd.DataFrame(result)
out_parquet = Path(OUT_DIR) / f"{RUN_ID}__kinematics_master.parquet"

# Ticket 004: Inject 4 per-row session label columns (UD-006 Option A: '651', 'T1', 'P2', 'R1').
# Parsed from RUN_ID stem; all values are str (object dtype); session-constant (same in every row).
import re as _re_t004
_label_re_t004 = _re_t004.compile(r'^(\d+)_(T\d+)_(P\d+)_(R\d+)')
_lm_t004 = _label_re_t004.match(str(RUN_ID))
if _lm_t004:
    df_master['subject_id'] = _lm_t004.group(1)
    df_master['timepoint']  = _lm_t004.group(2)
    df_master['piece']      = _lm_t004.group(3)
    df_master['rep']        = _lm_t004.group(4)
    print(f"[Ticket 004] Labels: subject_id={_lm_t004.group(1)}, timepoint={_lm_t004.group(2)}, "
          f"piece={_lm_t004.group(3)}, rep={_lm_t004.group(4)}")
else:
    import logging as _lg_t004a
    _lg_t004a.getLogger('ticket004').warning(
        "[Ticket 004] Could not parse session labels from RUN_ID=%s; setting all label columns to None.", RUN_ID
    )
    for _col_t004 in ('subject_id', 'timepoint', 'piece', 'rep'):
        df_master[_col_t004] = None
    print(f"[Ticket 004] WARNING: session labels not parsed from RUN_ID={RUN_ID}")

# Ticket 004: Read ref_is_fallback from S05 reference_metadata.json (Phase 11.5 R3 graceful handling).
_ref_meta_path_t004 = Path(DERIV_REF) / f"{RUN_ID}__reference_metadata.json"
try:
    with open(_ref_meta_path_t004) as _rf_t004:
        _ref_meta_data_t004 = json.load(_rf_t004)
    _rif_t004 = str(_ref_meta_data_t004.get('ref_is_fallback', 'null')).lower()
    if _rif_t004 not in ('true', 'false'):
        _rif_t004 = 'null'
    print(f"[Ticket 004] ref_is_fallback={_rif_t004}")
except Exception as _e_t004:
    import logging as _lg_t004b
    _lg_t004b.getLogger('ticket004').warning(
        "[Ticket 004] reference_metadata.json missing or unreadable for %s: %s; ref_is_fallback set to null.", RUN_ID, _e_t004
    )
    _rif_t004 = 'null'
    print(f"[Ticket 004] WARNING: ref_is_fallback=null (reference_metadata.json not found for {RUN_ID})")

# Ticket 007a: Read filter_psd_verdict from S04 filtering_summary.json (Phase 11.5 R3)
_t007a_psd_verdict = 'null'
try:
    _filt_sum_path = Path(DERIV_FILTERED) / f"{RUN_ID}__filtering_summary.json"
    with open(_filt_sum_path) as _f007a:
        _filt_sum = json.load(_f007a)
    _t007a_psd_verdict = str(
        _filt_sum.get('filter_params', {}).get('psd_audit', {}).get('session_psd_verdict') or 'null'
    )
    print(f"[Ticket 007a] filter_psd_verdict={_t007a_psd_verdict}")
except Exception as _e007a_psd:
    import logging as _lg007a
    _lg007a.getLogger('ticket007a').warning(
        "[Ticket 007a] filtering_summary.json missing for %s; filter_psd_verdict=null", RUN_ID
    )
    print(f"[Ticket 007a] WARNING: filter_psd_verdict=null (filtering_summary.json not found)")

# Ticket 007a: Read bone_qc_status from S02 preprocess_summary.json (Phase 11.5 R3)
_t007a_bone_qc = 'null'
try:
    _deriv_02_t007a = os.path.join(PROJECT_ROOT, CONFIG['derivatives_dir'], "step_02_preprocess")
    _preproc_sum_path = Path(_deriv_02_t007a) / f"{RUN_ID}__preprocess_summary.json"
    with open(_preproc_sum_path) as _f007a_b:
        _preproc_sum = json.load(_f007a_b)
    _t007a_bone_qc = str(_preproc_sum.get('bone_qc_status') or 'null')
    print(f"[Ticket 007a] bone_qc_status={_t007a_bone_qc}")
except Exception as _e007a_bone:
    import logging as _lg007a_b
    _lg007a_b.getLogger('ticket007a').warning(
        "[Ticket 007a] preprocess_summary.json missing for %s; bone_qc_status=null", RUN_ID
    )
    print(f"[Ticket 007a] WARNING: bone_qc_status=null (preprocess_summary.json not found)")

# --- NaN Integrity Guard (Enhanced v3.1) ---
# Threshold: > 0.1% triggers CRITICAL WARNING with offending column inventory
nan_counts = df_master.isna().sum()
total_cells = df_master.shape[0] * df_master.shape[1]
total_nans = int(nan_counts.sum())
nan_pct = (total_nans / total_cells) * 100 if total_cells > 0 else 0.0

# Collect offending columns for metadata injection
nan_offending_cols = {}
if total_nans > 0:
    nan_cols = nan_counts[nan_counts > 0].sort_values(ascending=False)
    for col_name, cnt in nan_cols.items():
        nan_offending_cols[col_name] = {
            "nan_count": int(cnt),
            "nan_pct": round(float(cnt / len(df_master) * 100), 4),
        }

if total_nans > 0:
    if nan_pct > 0.1:
        print(f"*** CRITICAL WARNING: {total_nans:,} NaN cells ({nan_pct:.3f}%) in df_master ***")
        print(f"    Final_NaN_Rescue_Pct > 0.1% — flagging export with WARNING")
        print(f"    Offending columns ({len(nan_offending_cols)}):")
        for col_name, info in list(nan_offending_cols.items())[:15]:
            print(f"      {col_name}: {info['nan_count']:,} ({info['nan_pct']:.2f}%)")
        if len(nan_offending_cols) > 15:
            print(f"      ... and {len(nan_offending_cols) - 15} more columns")
        print(f"    ACTION: Inspect Step 02 gapfill quality for this session")
        import logging as _lg
        _lg.getLogger(__name__).warning(
            "[NaN Guard] CRITICAL: %.3f%% NaN in df_master. Offending columns: %s",
            nan_pct, list(nan_offending_cols.keys())[:10],
        )
    else:
        print(f"[NaN Guard] Minor NaN count: {total_nans} ({nan_pct:.4f}%) — filling with linear interpolation")
        df_master = df_master.interpolate(method='linear', limit_direction='both')
        remaining = int(df_master.isna().sum().sum())
        if remaining > 0:
            df_master = df_master.fillna(method='ffill').fillna(method='bfill')
        print(f"    Remaining NaNs after fill: {int(df_master.isna().sum().sum())}")
else:
    print("[NaN Guard] PASS — zero NaN cells in df_master")

# --- Parquet Metadata Injection via PyArrow ---
subject_id = CONFIG.get('subject_id', 'unknown')
def _safe_float(v, default=0.0):
    try:
        return float(v)
    except (TypeError, ValueError):
        return default

subject_height_cm = str(CONFIG.get('subject_height_cm', '') or '')
subject_mass_kg = str(CONFIG.get('subject_mass_kg', '') or '')
pipeline_ver = CONFIG.get('pipeline_version', 'v3.0_com_enhanced')
proc_timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

custom_metadata = {
    b"run_id": RUN_ID.encode(),
    b"subject_id": str(subject_id).encode(),
    b"subject_height_cm": subject_height_cm.encode(),
    b"subject_mass_kg": subject_mass_kg.encode(),
    b"pipeline_version": str(pipeline_ver).encode(),
    b"processing_timestamp": proc_timestamp.encode(),
    b"com_segments_available": str(com_report['segments_available']).encode(),
    b"com_mass_coverage_pct": str(com_report['mass_available_pct']).encode(),
    b"nan_pct_before_guard": f"{nan_pct:.4f}".encode(),
    b"nan_guard_status": ("CRITICAL" if nan_pct > 0.1 else ("MINOR" if nan_pct > 0 else "CLEAN")).encode(),
    b"nan_offending_cols": json.dumps(list(nan_offending_cols.keys())[:20]).encode() if nan_offending_cols else b"[]",
    # Metadata quality sentinel
    b"metadata_quality": ("UNRELIABLE_COM_DEFAULT_ANTHRO"
                          if (abs(_safe_float(subject_height_cm) - 170.0) < 0.1 and abs(_safe_float(subject_mass_kg) - 70.0) < 0.1)
                          else ("MISSING_ANTHRO" if not subject_height_cm or not subject_mass_kg
                                else "SUBJECT_SPECIFIC")).encode(),
    # Phase 4 additions
    b"derivative_method": b"savgol_chunked_adaptive",
    b"chunking_guard": b"true",
    b"sg_window_len_global": str(W_LEN).encode(),
    b"sg_adapt_source": SG_ADAPT_SOURCE.encode(),
    b"sg_polyorder": str(SG_POLYORDER).encode(),
    b"subject_height_source": CONFIG.get('subject_height_source', 'unknown').encode(),
    b"com_reliability_score": str(com_reliability["com_reliability_score"]).encode(),
    b"com_reliability_flag": com_reliability["com_reliability_flag"].encode(),
    b"euler_standard": b"ISB",
    # Ticket 004: ref_is_fallback (already set above)
    b"ref_is_fallback": _rif_t004.encode(),
    # Ticket 007a: remaining 3 approved PyArrow metadata fields (Phase 11.5 R3 graceful handling)
    b"filter_psd_verdict": _t007a_psd_verdict.encode(),
    b"gate_01_status": b"PASS",  # Any session reaching S06 passed the S01 gate
    b"bone_qc_status": _t007a_bone_qc.encode(),
}

# --- Global Quaternion Continuity Enforcement ---
# Ensures no hemisphere flips survive into the final Parquet, even during
# 360-degree rotations across the full session.
# This is the FINAL pass and must cover ALL relative quaternion channels
# (both __qx/__qy/__qz/__qw and __rel_qx etc.) across the entire time-series.
import re as _re
quat_col_groups = {}
_quat_suffix_re = _re.compile(r'__q([xyzw])$')
for col in df_master.columns:
    m = _quat_suffix_re.search(col)
    if m:
        joint = col[:m.start()]
        quat_col_groups.setdefault(joint, []).append(col)

n_flips_fixed = 0
n_joints_enforced = 0
for joint, cols in quat_col_groups.items():
    if len(cols) != 4:
        continue
    cols_sorted = sorted(cols, key=lambda c: 'xyzw'.index(c[-2:].replace('q', '')))
    if not all(c in df_master.columns for c in cols_sorted):
        continue
    Q = df_master[cols_sorted].values.copy()
    flips_this_joint = 0
    for t in range(1, len(Q)):
        if np.dot(Q[t - 1], Q[t]) < 0:
            Q[t] *= -1.0
            flips_this_joint += 1
    if flips_this_joint > 0:
        n_flips_fixed += flips_this_joint
    df_master[cols_sorted] = Q
    n_joints_enforced += 1

if n_flips_fixed > 0:
    print(f"[Continuity] Fixed {n_flips_fixed} hemisphere flips across {n_joints_enforced} joint quaternion channels")
else:
    print(f"[Continuity] PASS — no hemisphere flips in {n_joints_enforced} quaternion channels")
print(f"[Continuity] Coverage: {n_joints_enforced} joints enforced (all relative quaternions in df_master)")

table = pa.Table.from_pandas(df_master, preserve_index=False)
merged_metadata = {**(table.schema.pandas_metadata or {})} if hasattr(table.schema, 'pandas_metadata') and table.schema.pandas_metadata else {}
existing_meta = table.schema.metadata or {}
existing_meta.update(custom_metadata)
table = table.replace_schema_metadata(existing_meta)
pq.write_table(table, str(out_parquet))

print(f"Saved: {out_parquet}")
print(f"  Metadata injected: run_id, subject_id, height_cm, mass_kg, pipeline_version, timestamp")
print(f"  Shape: {df_master.shape[0]:,} frames x {df_master.shape[1]} columns")

# ISB Euler sequences (from euler_isb.py, computed in Euler cell above)
euler_sequences_used = euler_audit["step_06_euler_sequences_used"]

# SavGol effective cutoff diagnostic (global fallback reference)
sg_eff_cutoff_hz = round((SG_POLYORDER + 1) / (3.2 * W_LEN) * FS, 2)

# Composite WBCoM bandwidth: mass-weighted harmonic mean of per-joint SavGol cutoffs
_composite_wbcom_bw = None
if SG_AUDIT:
    from com_engine import SEGMENT_PARAMS
    _seg_f3db = []
    _seg_mass = []
    for seg_label, seg_def in SEGMENT_PARAMS.items():
        prox = seg_def["proximal"]
        if prox in SG_AUDIT:
            _seg_f3db.append(SG_AUDIT[prox]["sg_f3db_hz"])
            _seg_mass.append(seg_def["mass_frac"])
    if _seg_f3db and sum(_seg_mass) > 0:
        _inv_sum = sum(m / f for m, f in zip(_seg_mass, _seg_f3db))
        _composite_wbcom_bw = round(sum(_seg_mass) / _inv_sum, 2) if _inv_sum > 0 else None

# Ticket 007b: Quaternion diagnostics (SO(3) integrity).
# Measured at two points to maximize scientific value:
#   - global_quat_integrity: raw __qx/__qy/__qz/__qw cols from S04 input (should be ~1.0)
#   - relative_quat_drift: __zeroed_rel_q* and __raw_rel_q* cols (NB06's own algebra; can drift)
# Renorm threshold per spec: |‖q‖ - 1| > 0.05 → counted toward renorm_burden_pct.
# Hemisphere flip count: re-uses n_flips_fixed from the continuity enforcement above.
import re as _re_t007b_q

def _t007b_quat_norm_stats(df_subset, joint, suffix_template):
    """Compute norm stats for one joint's 4 quaternion columns. Returns dict or None."""
    cols = [suffix_template.format(joint=joint, axis=a) for a in ('x', 'y', 'z', 'w')]
    if not all(c in df_subset.columns for c in cols):
        return None
    Q = df_subset[cols].values
    valid = ~np.isnan(Q).any(axis=1)
    Qv = Q[valid]
    if len(Qv) == 0:
        return None
    norms = np.linalg.norm(Qv, axis=1)
    rb = int(np.sum(np.abs(norms - 1.0) > 0.05))
    return {
        'quat_norm_mean': float(np.mean(norms)),
        'quat_norm_std':  float(np.std(norms)),
        'quat_norm_min':  float(np.min(norms)),
        'quat_norm_max':  float(np.max(norms)),
        'renorm_burden_pct': float(100.0 * rb / len(norms)),
        'n_frames_evaluated': int(len(norms)),
    }

_t007b_global_quat = {}
_t007b_rel_drift   = {}

# Discover joints from the global __qx columns present in df_in (S04 input).
# IMPORTANT: NB06 does NOT write the raw global __qx/__qy/__qz/__qw columns to
# df_master/kinematics_master.parquet — only the relative quaternions survive.
# So we measure global_quat_integrity from df_in (the S04 input read at the
# start of NB06). This confirms upstream pipeline integrity (S02/S03/S04
# normalization). The relative_quat_drift measurement (below) operates on
# df_master and captures NB06's own quaternion algebra drift.
_t007b_global_joints = sorted({
    c.replace('__qx', '') for c in df_in.columns
    if c.endswith('__qx') and 'rel' not in c
})
for _j_t007b in _t007b_global_joints:
    g = _t007b_quat_norm_stats(df_in, _j_t007b, '{joint}__q{axis}')
    if g is not None:
        _t007b_global_quat[_j_t007b] = g

# Discover joints from the zeroed_rel_qx columns
_t007b_rel_joints = sorted({
    c.replace('__zeroed_rel_qx', '') for c in df_master.columns
    if c.endswith('__zeroed_rel_qx')
})
for _j_t007b in _t007b_rel_joints:
    z = _t007b_quat_norm_stats(df_master, _j_t007b, '{joint}__zeroed_rel_q{axis}')
    if z is not None:
        _t007b_rel_drift[_j_t007b] = z

_t007b_renorm_thresh_pct = 5.0
_t007b_joints_above = [
    j for j, s in _t007b_rel_drift.items()
    if s['renorm_burden_pct'] > _t007b_renorm_thresh_pct
] + [
    j for j, s in _t007b_global_quat.items()
    if s['renorm_burden_pct'] > _t007b_renorm_thresh_pct and j not in _t007b_rel_drift
]
if _t007b_joints_above:
    _t007b_verdict = "REVIEW_SO3_SMOOTHING"
else:
    _t007b_verdict = "CURRENT_METHOD_ACCEPTABLE"

_t007b_quaternion_diagnostics = {
    'global_quat_integrity': _t007b_global_quat,
    'relative_quat_drift':   _t007b_rel_drift,
    'rotation_method_verdict': _t007b_verdict,
    'renorm_threshold_pct': _t007b_renorm_thresh_pct,
    'n_joints_above_renorm_threshold': len(_t007b_joints_above),
    'joints_above_renorm_threshold': _t007b_joints_above,
    'total_hemisphere_flips_corrected': int(n_flips_fixed),
    'n_joints_continuity_enforced': int(n_joints_enforced),
    'note': (
        'global_quat_integrity measures raw quaternion columns (post-S04 normalize, '
        'should be ~1.0). relative_quat_drift measures NB06-derived relative quaternions '
        'before the global continuity enforcement above. Hemisphere flips reported are '
        'detected-and-corrected by NB06 continuity loop.'
    ),
}
print(f"[Ticket 007b] quaternion_diagnostics: verdict={_t007b_verdict}, "
      f"flips_corrected={n_flips_fixed}, joints_above_renorm={len(_t007b_joints_above)}")

validation_report = {
    "run_id": RUN_ID,
    "total_frames": len(df_master),
    "quaternion_diagnostics": _t007b_quaternion_diagnostics,
    "lin_kine_diagnostics": _t007b_lin_kine_diagnostics,
    "per_joint": {r["joint"]: {k: v for k, v in r.items() if k != "joint"} for r in validation_rows},
    "per_segment_linear": {a["segment"]: {k: v for k, v in a.items() if k != "segment"} for a in lin_audit},
    "euler_sequences_used": euler_sequences_used,
    "euler_audit": {
        "isb_compliant": euler_audit["step_06_isb_compliant"],
        "non_isb_joints": euler_audit.get("step_06_non_isb_joints"),
    },
    "derivative_config": {
        "method": "savgol_chunked_adaptive",
        "sg_window_len_global": W_LEN,
        "sg_polyorder": SG_POLYORDER,
        "sg_effective_cutoff_hz_global": sg_eff_cutoff_hz,
        "lin_acc_method": "savgol_deriv2_from_position",
        "ang_acc_method": "savgol_deriv1_from_omega",
        "chunking_guard": True,
    },
    "savgol_adaptation": {
        "mode": SG_ADAPT_SOURCE,
        "target_multiplier": 1.2,
        "floor_w": 7,
        "ceiling_w": W_LEN,
        "fallback_w": W_LEN,
    },
    "per_joint_savgol_audit": SG_AUDIT if SG_AUDIT else {"note": "global_fallback, no per-joint data"},
    "composite_wbcom_bandwidth_hz": _composite_wbcom_bw,
    "com_reliability": com_reliability,
}
# Ticket 007b: sanitize float NaN/Inf to None before JSON write (identical pattern to Ticket 005).
# Pre-existing NB06 bug: per_segment_linear contained literal `NaN` floats from SavGol all-NaN
# derivative outputs for joints whose position columns survived the gate but contained all-NaN
# values. Default json.dump emits invalid JSON `NaN` literal. None serializes to JSON `null` (valid).
def _t007b_sanitize_nan(obj):
    import math as _math_t007b
    if isinstance(obj, dict):
        return {k: _t007b_sanitize_nan(v) for k, v in obj.items()}
    if isinstance(obj, (list, tuple)):
        return [_t007b_sanitize_nan(v) for v in obj]
    if isinstance(obj, float) and not _math_t007b.isfinite(obj):
        return None
    return obj

out_json = Path(OUT_DIR) / f"{RUN_ID}__validation_report.json"
with open(out_json, "w", encoding="utf-8") as f:
    json.dump(_t007b_sanitize_nan(validation_report), f, indent=2, allow_nan=False)
print(f"Saved: {out_json}")
max_geodesic = max(r["geodesic_offset_std"] for r in validation_rows)
mean_align = np.mean([r["velocity_alignment_pct"] for r in validation_rows])
print("Velocity Magnitude Alignment: mean similarity (raw vs zeroed) = {:.2f}% (expect close to 100%).".format(mean_align))
print("Stability Proof (quaternion): max Std of geodesic distance over joints = {:.6f}° (expect near zero; constant T-pose → constant rotation).".format(max_geodesic))
print()
validation_table = pd.DataFrame([{"joint": r["joint"], "velocity_similarity_pct": r["velocity_alignment_pct"], "geodesic_std_deg": r["geodesic_offset_std"]} for r in validation_rows])
print("Validation Summary Table (quaternion-based proof for proceeding with zeroed data):")
display(validation_table)
if mean_align >= 99.0 and np.isfinite(max_geodesic) and max_geodesic < 0.1:
    print("Conclusion: T-pose is stable; zeroed data aligns with raw. Proceed using zeroed data.")
else:
    print("Conclusion: Review T-pose or per-joint metrics above before relying solely on zeroed data.")
```

```python
total_frames = len(df_master)
THRESH = {
    'rotation_mag_deg': {'WARNING': 140.0, 'ALERT': 160.0, 'CRITICAL': 180.0},
    'angular_velocity_deg_s': {'WARNING': 800.0, 'ALERT': 1200.0, 'CRITICAL': 1500.0},
    'angular_acceleration_deg_s2': {'WARNING': 35000.0, 'ALERT': 50000.0, 'CRITICAL': 80000.0},
    'linear_velocity_mm_s': {'WARNING': 3000.0, 'ALERT': 5000.0, 'CRITICAL': 7000.0},
    'linear_acceleration_mm_s2': {'WARNING': 60000.0, 'ALERT': 100000.0, 'CRITICAL': 150000.0},
}

def outlier_table(joint_or_segment_list, magnitude_per_name, thresh_dict, total_frames, name_col='joint'):
    rows = []
    for name in joint_or_segment_list:
        mag = magnitude_per_name.get(name)
        if mag is None or len(mag) == 0:
            continue
        n_w = int(np.sum(mag > thresh_dict['WARNING']))
        n_a = int(np.sum(mag > thresh_dict['ALERT']))
        n_c = int(np.sum(mag > thresh_dict['CRITICAL']))
        rows.append({
            name_col: name,
            'n_frames_WARNING': n_w, 'n_frames_ALERT': n_a, 'n_frames_CRITICAL': n_c,
            'pct_WARNING': round(100.0 * n_w / total_frames, 2),
            'pct_ALERT': round(100.0 * n_a / total_frames, 2),
            'pct_CRITICAL': round(100.0 * n_c / total_frames, 2),
            'total_frames': total_frames,
        })
    return pd.DataFrame(rows)

# 1. Rotation vector magnitude (total rotation from T-pose) — zeroed quat
rot_mag_per_joint = {}
for j in kinematics_map:
    qc = [f"{j}__zeroed_rel_qx", f"{j}__zeroed_rel_qy", f"{j}__zeroed_rel_qz", f"{j}__zeroed_rel_qw"]
    if not all(c in df_master.columns for c in qc):
        continue
    q = df_master[qc].values
    rot_mag_per_joint[j] = np.degrees(np.linalg.norm(R.from_quat(q).as_rotvec(), axis=1))
df_rot = outlier_table(list(rot_mag_per_joint.keys()), rot_mag_per_joint, THRESH['rotation_mag_deg'], total_frames)
print("Rotation magnitude (deg) — thresholds 140° / 160° / 180°")
display(df_rot)

# 2. Angular velocity (deg/s)
omega_mag_per_joint = {}
for j in kinematics_map:
    cols = [f"{j}__zeroed_rel_omega_x", f"{j}__zeroed_rel_omega_y", f"{j}__zeroed_rel_omega_z"]
    if not all(c in df_master.columns for c in cols):
        continue
    omega_mag_per_joint[j] = np.linalg.norm(df_master[cols].values, axis=1)
df_omega = outlier_table(list(omega_mag_per_joint.keys()), omega_mag_per_joint, THRESH['angular_velocity_deg_s'], total_frames)
print("Angular velocity (deg/s) — thresholds 800 / 1200 / 1500")
display(df_omega)

# 3. Angular acceleration (deg/s²)
alpha_mag_per_joint = {}
for j in kinematics_map:
    cols = [f"{j}__zeroed_rel_alpha_x", f"{j}__zeroed_rel_alpha_y", f"{j}__zeroed_rel_alpha_z"]
    if not all(c in df_master.columns for c in cols):
        continue
    alpha_mag_per_joint[j] = np.linalg.norm(df_master[cols].values, axis=1)
df_alpha = outlier_table(list(alpha_mag_per_joint.keys()), alpha_mag_per_joint, THRESH['angular_acceleration_deg_s2'], total_frames)
print("Angular acceleration (deg/s²) — thresholds 35000 / 50000 / 80000")
display(df_alpha)

# 4. Linear velocity (mm/s) — per segment
lin_vel_segments = [c.split('__')[0] for c in df_master.columns if c.endswith('__lin_vel_rel_x')]
lin_vel_mag = {}
for seg in lin_vel_segments:
    cx, cy, cz = f"{seg}__lin_vel_rel_x", f"{seg}__lin_vel_rel_y", f"{seg}__lin_vel_rel_z"
    if cx in df_master.columns and cy in df_master.columns and cz in df_master.columns:
        lin_vel_mag[seg] = np.linalg.norm(df_master[[cx, cy, cz]].values, axis=1)
df_linvel = outlier_table(list(lin_vel_mag.keys()), lin_vel_mag, THRESH['linear_velocity_mm_s'], total_frames, name_col='segment')
print("Linear velocity (mm/s) — thresholds 3000 / 5000 / 7000")
display(df_linvel)

# 5. Linear acceleration (mm/s²)
lin_acc_segments = [c.split('__')[0] for c in df_master.columns if c.endswith('__lin_acc_rel_x')]
lin_acc_mag = {}
for seg in lin_acc_segments:
    cx, cy, cz = f"{seg}__lin_acc_rel_x", f"{seg}__lin_acc_rel_y", f"{seg}__lin_acc_rel_z"
    if cx in df_master.columns and cy in df_master.columns and cz in df_master.columns:
        lin_acc_mag[seg] = np.linalg.norm(df_master[[cx, cy, cz]].values, axis=1)
df_linacc = outlier_table(list(lin_acc_mag.keys()), lin_acc_mag, THRESH['linear_acceleration_mm_s2'], total_frames, name_col='segment')
print("Linear acceleration (mm/s²) — thresholds 60000 / 100000 / 150000")
display(df_linacc)

# --- Surgical repair (when ENFORCE_CLEANING and any Critical): delegate to kinematic_repair ---
from kinematic_repair import identify_critical_units, apply_surgical_repair

angular_critical_joints, linear_critical_segments = identify_critical_units(
    df_rot, df_omega, df_alpha, df_linvel, df_linacc
)

if ENFORCE_CLEANING and (angular_critical_joints or linear_critical_segments):
    apply_surgical_repair(
        result,
        pos_rel,
        validation_rows,
        lin_audit,
        angular_critical_joints,
        linear_critical_segments,
        rot_mag_per_joint,
        omega_mag_per_joint,
        alpha_mag_per_joint,
        lin_vel_mag,
        lin_acc_mag,
        THRESH,
        kinematics_map,
        ref_pose,
        FS,
        W_LEN,
        SG_POLYORDER,
        dt,
        OMEGA_THRESH,
        LIN_ACC_THRESH,
        w_len_map=W_LEN_MAP,
    )
    # (Angular/linear repair done in-place by kinematic_repair; now save and rebuild outlier tables)
    df_master = pd.DataFrame(result)
    out_parquet = Path(OUT_DIR) / f"{RUN_ID}__kinematics_master.parquet"

    # NaN guard (post-repair)
    nan_counts_r = df_master.isna().sum()
    total_cells_r = df_master.shape[0] * df_master.shape[1]
    total_nans_r = int(nan_counts_r.sum())
    nan_pct_r = (total_nans_r / total_cells_r) * 100 if total_cells_r > 0 else 0.0
    if total_nans_r > 0:
        nan_cols_r = nan_counts_r[nan_counts_r > 0].sort_values(ascending=False)
        if nan_pct_r > 0.1:
            print(f"*** CRITICAL WARNING (post-repair): {total_nans_r:,} NaN cells ({nan_pct_r:.3f}%) ***")
            for c_name, cnt in nan_cols_r.head(5).items():
                print(f"      {c_name}: {cnt:,}")
        else:
            df_master = df_master.interpolate(method='linear', limit_direction='both')
            df_master = df_master.fillna(method='ffill').fillna(method='bfill')
            print(f"[NaN Guard] Minor NaN ({nan_pct_r:.4f}%) filled via interpolation")

    # --- Global Quaternion Continuity (post-repair) ---
    for joint_r, cols_r in quat_col_groups.items():
        if len(cols_r) != 4:
            continue
        cols_s = sorted(cols_r, key=lambda c: 'xyzw'.index(c[-1]))
        if not all(c in df_master.columns for c in cols_s):
            continue
        Q_r = df_master[cols_s].values.copy()
        for t_r in range(1, len(Q_r)):
            if np.dot(Q_r[t_r - 1], Q_r[t_r]) < 0:
                Q_r[t_r] *= -1.0
        df_master[cols_s] = Q_r

    # Re-inject metadata
    table_r = pa.Table.from_pandas(df_master, preserve_index=False)
    meta_r = table_r.schema.metadata or {}
    meta_r.update(custom_metadata)
    meta_r[b"processing_timestamp"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S").encode()
    table_r = table_r.replace_schema_metadata(meta_r)
    pq.write_table(table_r, str(out_parquet))
    print("Surgical repair applied; overwrote", out_parquet, "(with metadata)")
    validation_report = {"run_id": RUN_ID, "total_frames": len(df_master), "per_joint": {r["joint"]: {k: v for k, v in r.items() if k != "joint"} for r in validation_rows}, "per_segment_linear": {a["segment"]: {k: v for k, v in a.items() if k != "segment"} for a in lin_audit}}
    out_json = Path(OUT_DIR) / f"{RUN_ID}__validation_report.json"
    with open(out_json, "w", encoding="utf-8") as f:
        json.dump(validation_report, f, indent=2)
    print("Overwrote", out_json)
    # Rebuild outlier tables from repaired df_master
    rot_mag_per_joint = {j: np.degrees(np.linalg.norm(R.from_quat(df_master[[f"{j}__zeroed_rel_qx", f"{j}__zeroed_rel_qy", f"{j}__zeroed_rel_qz", f"{j}__zeroed_rel_qw"]].values).as_rotvec(), axis=1)) for j in kinematics_map if all(f"{j}__zeroed_rel_q{c}" in df_master.columns for c in "xyzw")}
    omega_mag_per_joint = {j: np.linalg.norm(df_master[[f"{j}__zeroed_rel_omega_x", f"{j}__zeroed_rel_omega_y", f"{j}__zeroed_rel_omega_z"]].values, axis=1) for j in kinematics_map if all(f"{j}__zeroed_rel_omega_{c}" in df_master.columns for c in "xyz")}
    alpha_mag_per_joint = {j: np.linalg.norm(df_master[[f"{j}__zeroed_rel_alpha_x", f"{j}__zeroed_rel_alpha_y", f"{j}__zeroed_rel_alpha_z"]].values, axis=1) for j in kinematics_map if all(f"{j}__zeroed_rel_alpha_{c}" in df_master.columns for c in "xyz")}
    lin_vel_mag = {seg: np.linalg.norm(df_master[[f"{seg}__lin_vel_rel_x", f"{seg}__lin_vel_rel_y", f"{seg}__lin_vel_rel_z"]].values, axis=1) for seg in lin_vel_segments if all(f"{seg}__lin_vel_rel_{c}" in df_master.columns for c in "xyz")}
    lin_acc_mag = {seg: np.linalg.norm(df_master[[f"{seg}__lin_acc_rel_x", f"{seg}__lin_acc_rel_y", f"{seg}__lin_acc_rel_z"]].values, axis=1) for seg in lin_acc_segments if all(f"{seg}__lin_acc_rel_{c}" in df_master.columns for c in "xyz")}
    df_rot = outlier_table(list(rot_mag_per_joint.keys()), rot_mag_per_joint, THRESH["rotation_mag_deg"], total_frames)
    df_omega = outlier_table(list(omega_mag_per_joint.keys()), omega_mag_per_joint, THRESH["angular_velocity_deg_s"], total_frames)
    df_alpha = outlier_table(list(alpha_mag_per_joint.keys()), alpha_mag_per_joint, THRESH["angular_acceleration_deg_s2"], total_frames)
    df_linvel = outlier_table(list(lin_vel_mag.keys()), lin_vel_mag, THRESH["linear_velocity_mm_s"], total_frames, name_col="segment")
    df_linacc = outlier_table(list(lin_acc_mag.keys()), lin_acc_mag, THRESH["linear_acceleration_mm_s2"], total_frames, name_col="segment")

# ============================================================
# PHASE 2: PATH LENGTH & BILATERAL SYMMETRY COMPUTATION
# ============================================================
# Must run BEFORE validation_report export (moved here from cells 15-16)

def compute_path_length(positions_mm):
    """Compute cumulative 3D path length from position time series."""
    if len(positions_mm) < 2:
        return 0.0
    deltas = np.diff(positions_mm, axis=0)
    distances = np.linalg.norm(deltas, axis=1)
    total_mm = np.sum(distances)
    total_m = total_mm / 1000.0
    return total_m

def compute_bilateral_symmetry(left_values, right_values, metric_name="metric"):
    """Compute bilateral symmetry index between paired measurements."""
    left = float(left_values)
    right = float(right_values)
    if left == 0 and right == 0:
        return {"left_value": 0.0, "right_value": 0.0, "absolute_diff": 0.0, "percent_diff": 0.0, "symmetry_index": 1.0, "metric_name": metric_name}
    abs_diff = abs(left - right)
    max_val = max(abs(left), abs(right))
    if max_val == 0:
        symmetry_index = 1.0
        percent_diff = 0.0
    else:
        symmetry_index = 1.0 - (abs_diff / max_val)
        percent_diff = (abs_diff / max_val) * 100.0
    return {"left_value": left, "right_value": right, "absolute_diff": abs_diff, "percent_diff": percent_diff, "symmetry_index": symmetry_index, "metric_name": metric_name}

# Compute path length for all segments
print("\n" + "="*80)
print("PHASE 2: Computing path length and bilateral symmetry...")
print("="*80)

path_lengths = {}
for segment in kinematics_map:
    pos_cols = [f"{segment}__lin_rel_px", f"{segment}__lin_rel_py", f"{segment}__lin_rel_pz"]
    if all(col in df_master.columns for col in pos_cols):
        positions = df_master[pos_cols].values
        path_length_m = compute_path_length(positions)
        path_lengths[segment] = path_length_m
    else:
        path_lengths[segment] = 0.0

path_lengths_sorted = dict(sorted(path_lengths.items(), key=lambda x: x[1], reverse=True))
print(f"✓ Path length computed for {len(path_lengths)} segments")
print(f"  Top 3: {list(path_lengths_sorted.keys())[:3]}")

# Compute Intensity Index (Phase 3)
# Intensity = path_length / duration (m/s, average velocity-like measure)
duration_sec = n_frames / FS
intensity_index = {}
for segment, path_m in path_lengths.items():
    if duration_sec > 0:
        intensity_index[segment] = path_m / duration_sec  # m/s
    else:
        intensity_index[segment] = 0.0

intensity_sorted = dict(sorted(intensity_index.items(), key=lambda x: x[1], reverse=True))
print(f"✓ Intensity Index computed (path/duration in m/s)")
print(f"  Most intense: {list(intensity_sorted.keys())[:3]}")

# Compute bilateral symmetry
BILATERAL_PAIRS = {
    "upper_arm": ("LeftArm", "RightArm"),
    "forearm": ("LeftForeArm", "RightForeArm"),
    "hand": ("LeftHand", "RightHand"),
    "thigh": ("LeftUpLeg", "RightUpLeg"),
    "shin": ("LeftLeg", "RightLeg"),
    "foot": ("LeftFoot", "RightFoot"),
}

bilateral_symmetry = {}

# Path length symmetry
for pair_name, (left_seg, right_seg) in BILATERAL_PAIRS.items():
    if left_seg in path_lengths and right_seg in path_lengths:
        symmetry = compute_bilateral_symmetry(path_lengths[left_seg], path_lengths[right_seg], metric_name=f"path_length_{pair_name}")
        bilateral_symmetry[f"path_length_{pair_name}"] = symmetry

# Angular velocity symmetry
for pair_name, (left_seg, right_seg) in BILATERAL_PAIRS.items():
    if left_seg in omega_mag_per_joint and right_seg in omega_mag_per_joint:
        left_max = np.max(omega_mag_per_joint[left_seg])
        right_max = np.max(omega_mag_per_joint[right_seg])
        symmetry = compute_bilateral_symmetry(left_max, right_max, metric_name=f"max_omega_{pair_name}")
        bilateral_symmetry[f"max_omega_{pair_name}"] = symmetry

# Linear acceleration symmetry
for pair_name, (left_seg, right_seg) in BILATERAL_PAIRS.items():
    if left_seg in lin_acc_mag and right_seg in lin_acc_mag:
        left_max = np.max(lin_acc_mag[left_seg])
        right_max = np.max(lin_acc_mag[right_seg])
        symmetry = compute_bilateral_symmetry(left_max, right_max, metric_name=f"max_lin_acc_{pair_name}")
        bilateral_symmetry[f"max_lin_acc_{pair_name}"] = symmetry

print(f"✓ Bilateral symmetry computed for {len(bilateral_symmetry)} metrics")
print("="*80 + "\n")

# ============================================================
# 3-TIER BURST ANALYSIS & DATA RETENTION
# ============================================================
print("\n" + "="*80)
print("BURST ANALYSIS: Classifying Outlier Events")
print("="*80)

# Create frame-by-frame outlier mask (any metric exceeds WARNING threshold)
outlier_mask = np.zeros(total_frames, dtype=bool)

# Mark frames where ANY joint exceeds WARNING threshold in angular velocity
for joint, omega_vals in omega_mag_per_joint.items():
    outlier_mask |= (omega_vals >= THRESH['angular_velocity_deg_s']['WARNING'])

# Mark frames where ANY joint exceeds WARNING in angular acceleration
for joint, alpha_vals in alpha_mag_per_joint.items():
    outlier_mask |= (alpha_vals >= THRESH['angular_acceleration_deg_s2']['WARNING'])

# Mark frames where ANY segment exceeds WARNING in linear acceleration
for seg, acc_vals in lin_acc_mag.items():
    outlier_mask |= (acc_vals >= THRESH['linear_acceleration_mm_s2']['WARNING'])

# Find consecutive runs of outliers
def find_consecutive_runs(mask):
    """Find start, end, and length of consecutive True runs in boolean mask."""
    runs = []
    in_run = False
    start = 0
    
    for i, val in enumerate(mask):
        if val and not in_run:
            # Start of new run
            start = i
            in_run = True
        elif not val and in_run:
            # End of run
            runs.append({'start': start, 'end': i-1, 'length': i - start})
            in_run = False
    
    # Handle case where run extends to end
    if in_run:
        runs.append({'start': start, 'end': len(mask)-1, 'length': len(mask) - start})
    
    return runs

outlier_runs = find_consecutive_runs(outlier_mask)

# Classify into 3 tiers
tier1_artifacts = [r for r in outlier_runs if 1 <= r['length'] <= 3]
tier2_bursts = [r for r in outlier_runs if 4 <= r['length'] <= 7]
tier3_flows = [r for r in outlier_runs if r['length'] >= 8]

# Count artifact frames (Tier 1 only - these are excluded)
artifact_frames = sum(r['length'] for r in tier1_artifacts)
artifact_rate = (artifact_frames / total_frames * 100) if total_frames > 0 else 0.0

# Data retention (frames NOT in Tier 1 artifacts)
retained_frames = total_frames - artifact_frames
retention_rate = (retained_frames / total_frames * 100) if total_frames > 0 else 100.0

# Max consecutive outlier run
max_consecutive = max([r['length'] for r in outlier_runs]) if outlier_runs else 0

print(f"Total outlier events: {len(outlier_runs)}")
print(f"  Tier 1 (Artifacts, 1-3 frames): {len(tier1_artifacts)} events, {artifact_frames} frames ({artifact_rate:.3f}%)")
print(f"  Tier 2 (Bursts, 4-7 frames): {len(tier2_bursts)} events")
print(f"  Tier 3 (Flows, 8+ frames): {len(tier3_flows)} events")
print(f"  Max consecutive outlier run: {max_consecutive} frames")
print(f"  Data retention: {retention_rate:.2f}% ({retained_frames}/{total_frames} frames)")

# Compute clean statistics (excluding Tier 1 artifacts)
clean_mask = ~outlier_mask.copy()
for run in tier1_artifacts:
    clean_mask[run['start']:run['end']+1] = True  # Keep Tier 2&3 as they're legitimate movement

# Actually, let's exclude ALL outlier frames for clean stats
clean_mask = ~outlier_mask

# Compute clean angular velocity statistics
clean_omega_values = []
for joint, omega_vals in omega_mag_per_joint.items():
    clean_omega_values.extend(omega_vals[clean_mask].tolist())

if clean_omega_values:
    clean_max_vel = max(clean_omega_values)
    clean_mean_vel = np.mean(clean_omega_values)
else:
    clean_max_vel = 0.0
    clean_mean_vel = 0.0

# Original (with outliers) max for comparison
raw_max_vel = max([np.max(vals) for vals in omega_mag_per_joint.values()]) if omega_mag_per_joint else 0.0
velocity_reduction = ((raw_max_vel - clean_max_vel) / raw_max_vel * 100) if raw_max_vel > 0 else 0.0

print(f"\nClean velocity statistics (outliers excluded):")
print(f"  Raw max: {raw_max_vel:.2f} deg/s")
print(f"  Clean max: {clean_max_vel:.2f} deg/s")
print(f"  Clean mean: {clean_mean_vel:.2f} deg/s")
print(f"  Velocity reduction: {velocity_reduction:.2f}%")
print("="*80 + "\n")

# ============================================================
# OUTLIER VALIDATION & VALIDATION REPORT EXPORT
# ============================================================

# Save outlier validation tables to JSON (standalone and into validation_report)
outlier_report = {
    "run_id": RUN_ID,
    "total_frames": total_frames,
    "thresholds": THRESH,
    "rotation_mag_deg": df_rot.to_dict(orient="records"),
    "angular_velocity_deg_s": df_omega.to_dict(orient="records"),
    "angular_acceleration_deg_s2": df_alpha.to_dict(orient="records"),
    "linear_velocity_mm_s": df_linvel.to_dict(orient="records"),
    "linear_acceleration_mm_s2": df_linacc.to_dict(orient="records"),
}
outlier_json_path = Path(OUT_DIR) / f"{RUN_ID}__outlier_validation.json"
with open(outlier_json_path, "w", encoding="utf-8") as f:
    json.dump(outlier_report, f, indent=2)
print(f"Saved: {outlier_json_path}")

# Append outlier_validation (frame counts and percentages) to validation_report.json
validation_report_path = Path(OUT_DIR) / f"{RUN_ID}__validation_report.json"
if validation_report_path.exists():
    with open(validation_report_path, "r", encoding="utf-8") as f:
        v = json.load(f)
    v["total_frames"] = total_frames
    v["outlier_validation"] = {
        "total_frames": total_frames,
        "thresholds": THRESH,
        "rotation_mag_deg": df_rot.to_dict(orient="records"),
        "angular_velocity_deg_s": df_omega.to_dict(orient="records"),
        "angular_acceleration_deg_s2": df_alpha.to_dict(orient="records"),
        "linear_velocity_mm_s": df_linvel.to_dict(orient="records"),
        "linear_acceleration_mm_s2": df_linacc.to_dict(orient="records"),
    }
    
    # Compute aggregate outlier statistics from outlier_validation arrays
    # Sum up all WARNING+ALERT+CRITICAL frames across all metrics and joints
    total_outlier_frames = 0
    max_consecutive = 0
    
    # Count outliers from each metric category
    for metric_name in ["angular_velocity_deg_s", "angular_acceleration_deg_s2", "linear_acceleration_mm_s2"]:
        if metric_name in v["outlier_validation"]:
            for joint_data in v["outlier_validation"][metric_name]:
                if isinstance(joint_data, dict):
                    # Sum WARNING + ALERT + CRITICAL frames for this joint/metric
                    joint_outliers = (
                        joint_data.get("n_frames_WARNING", 0) + 
                        joint_data.get("n_frames_ALERT", 0) + 
                        joint_data.get("n_frames_CRITICAL", 0)
                    )
                    # Note: We take the max across metrics for a given frame, 
                    # but this is an approximation (assumes outliers don't overlap perfectly)
                    if joint_outliers > total_outlier_frames:
                        total_outlier_frames = joint_outliers
    
    outlier_percent = (total_outlier_frames / total_frames * 100) if total_frames > 0 else 0.0
    
    # Add aggregate outlier statistics
    v["outlier_analysis"] = {
        "counts": {
            "total_outliers": total_outlier_frames
        },
        "percentages": {
            "total_outliers": outlier_percent
        },
        "consecutive_runs": {
            "max_consecutive_any_outlier": max_consecutive
        }
    }
    
    # Add 3-tier burst analysis
    v["step_06_burst_analysis"] = {
        "classification": {
            "artifact_count": len(tier1_artifacts),
            "burst_count": len(tier2_bursts),
            "flow_count": len(tier3_flows),
            "total_events": len(outlier_runs)
        },
        "frame_statistics": {
            "artifact_frames": artifact_frames,
            "artifact_rate_percent": round(artifact_rate, 4),
            "retained_frames": retained_frames,
            "retention_rate_percent": round(retention_rate, 4)
        }
    }
    
    # Add clean velocity statistics
    v["clean_statistics"] = {
        "clean_statistics": {
            "max_deg_s": round(clean_max_vel, 2),
            "mean_deg_s": round(clean_mean_vel, 2)
        },
        "comparison": {
            "raw_max_deg_s": round(raw_max_vel, 2),
            "velocity_reduction_percent": round(velocity_reduction, 2)
        },
        "data_retention": {
            "frames_excluded": artifact_frames,
            "frames_retained": retained_frames,
            "retention_percent": round(retention_rate, 4)
        }
    }
    
    # Phase 2: Add path length and bilateral symmetry metrics
    v["path_length_m"] = path_lengths_sorted
    v["bilateral_symmetry"] = bilateral_symmetry
    
    # Phase 3: Add intensity index
    v["intensity_index_m_per_s"] = intensity_sorted
    
    # Phase 2.5: Group path lengths by anatomical region
    # Mapping: which segments contribute to each anatomical region
    region_mapping = {
        "Neck": ["Neck"],
        "Shoulders": ["LeftShoulder", "RightShoulder"],  # max of left/right
        "Elbows": ["LeftArm", "RightArm"],  # forearm segments (max of left/right)
        "Wrists": ["LeftForeArm", "RightForeArm"],  # wrist/hand segments (max of left/right)
        "Spine": ["Spine", "Spine1"],  # mid-back/thoracic (sum or max)
        "Hips": ["Hips", "LeftUpLeg", "RightUpLeg"],  # pelvis + hip joints (max)
        "Knees": ["LeftLeg", "RightLeg"],  # shin segments (max of left/right)
        "Ankles": ["LeftFoot", "RightFoot"],  # foot/ankle joints (max of left/right)
    }
    
    path_by_region = {}
    intensity_by_region = {}
    
    for region, segments in region_mapping.items():
        # Get path lengths for all segments in this region
        region_paths = [path_lengths.get(seg, 0.0) for seg in segments]
        region_intensities = [intensity_index.get(seg, 0.0) for seg in segments]
        
        # Use max for bilateral pairs, sum for axial structures
        if region in ["Neck", "Spine", "Hips"]:
            # For axial structures, use max (most active segment represents the region)
            path_by_region[region] = max(region_paths) if region_paths else 0.0
            intensity_by_region[region] = max(region_intensities) if region_intensities else 0.0
        else:
            # For bilateral structures, use max (most active side)
            path_by_region[region] = max(region_paths) if region_paths else 0.0
            intensity_by_region[region] = max(region_intensities) if region_intensities else 0.0
    
    # Create nested structure expected by notebook 08
    v["path_length"] = {
        "by_segment": path_lengths_sorted,
        "by_region": path_by_region
    }
    
    v["intensity_index"] = {
        "by_segment": intensity_sorted,
        "by_region": intensity_by_region
    }
    
    with open(validation_report_path, "w", encoding="utf-8") as f:
        json.dump(v, f, indent=2)
    print(f"Updated: {validation_report_path} (outlier + path_length + bilateral_symmetry + intensity_index + regional_grouping)")
```

## Phase 2: Path Length & Bilateral Symmetry

Compute cumulative path length (movement intensity) and bilateral symmetry indices for paired limbs.

```python
# ============================================================
# PATH LENGTH COMPUTATION (Movement Intensity Index)
# ============================================================

def compute_path_length(positions_mm):
    """
    Compute cumulative 3D path length from position time series.
    
    Args:
        positions_mm: (T, 3) array of [x, y, z] positions in mm
        
    Returns:
        float: Total path length in meters
    """
    if len(positions_mm) < 2:
        return 0.0
    
    # Compute frame-to-frame Euclidean distances
    deltas = np.diff(positions_mm, axis=0)
    distances = np.linalg.norm(deltas, axis=1)
    
    # Sum and convert mm -> meters
    total_mm = np.sum(distances)
    total_m = total_mm / 1000.0
    
    return total_m

# Compute path length for all segments
print("\nComputing path length (movement intensity)...")
path_lengths = {}

for segment in kinematics_map:
    pos_cols = [f"{segment}__lin_rel_px", f"{segment}__lin_rel_py", f"{segment}__lin_rel_pz"]
    
    if all(col in df_master.columns for col in pos_cols):
        positions = df_master[pos_cols].values
        path_length_m = compute_path_length(positions)
        path_lengths[segment] = path_length_m
    else:
        path_lengths[segment] = 0.0

# Sort by path length (most active segments first)
path_lengths_sorted = dict(sorted(path_lengths.items(), key=lambda x: x[1], reverse=True))

print(f"\n✓ Path length computed for {len(path_lengths)} segments")
print(f"  Most active: {list(path_lengths_sorted.keys())[:5]}")
print(f"  Range: {min(path_lengths.values()):.2f}m - {max(path_lengths.values()):.2f}m")
```

```python
# ============================================================
# BILATERAL SYMMETRY COMPUTATION
# ============================================================

def compute_bilateral_symmetry(left_values, right_values, metric_name="metric"):
    """
    Compute bilateral symmetry index between paired left/right measurements.
    
    Symmetry Index = 1 - |L - R| / max(L, R)
    
    Returns:
        dict: {
            "left_value": float,
            "right_value": float,
            "absolute_diff": float,
            "percent_diff": float,  # Relative difference (%)
            "symmetry_index": float,  # 1.0 = perfect symmetry, 0.0 = complete asymmetry
            "metric_name": str
        }
    """
    left = float(left_values)
    right = float(right_values)
    
    if left == 0 and right == 0:
        return {
            "left_value": 0.0,
            "right_value": 0.0,
            "absolute_diff": 0.0,
            "percent_diff": 0.0,
            "symmetry_index": 1.0,  # Perfect symmetry (both zero)
            "metric_name": metric_name
        }
    
    abs_diff = abs(left - right)
    max_val = max(abs(left), abs(right))
    
    if max_val == 0:
        symmetry_index = 1.0
        percent_diff = 0.0
    else:
        symmetry_index = 1.0 - (abs_diff / max_val)
        percent_diff = (abs_diff / max_val) * 100.0
    
    return {
        "left_value": left,
        "right_value": right,
        "absolute_diff": abs_diff,
        "percent_diff": percent_diff,
        "symmetry_index": symmetry_index,
        "metric_name": metric_name
    }

# Define bilateral pairs (standard anatomical naming)
BILATERAL_PAIRS = {
    "upper_arm": ("LeftArm", "RightArm"),
    "forearm": ("LeftForeArm", "RightForeArm"),
    "hand": ("LeftHand", "RightHand"),
    "thigh": ("LeftUpLeg", "RightUpLeg"),
    "shin": ("LeftLeg", "RightLeg"),
    "foot": ("LeftFoot", "RightFoot"),
}

print("\nComputing bilateral symmetry...")
bilateral_symmetry = {}

# Path length symmetry
for pair_name, (left_seg, right_seg) in BILATERAL_PAIRS.items():
    if left_seg in path_lengths and right_seg in path_lengths:
        symmetry = compute_bilateral_symmetry(
            path_lengths[left_seg],
            path_lengths[right_seg],
            metric_name=f"path_length_{pair_name}"
        )
        bilateral_symmetry[f"path_length_{pair_name}"] = symmetry

# Angular velocity symmetry (max omega)
for pair_name, (left_seg, right_seg) in BILATERAL_PAIRS.items():
    if left_seg in omega_mag_per_joint and right_seg in omega_mag_per_joint:
        left_max = np.max(omega_mag_per_joint[left_seg])
        right_max = np.max(omega_mag_per_joint[right_seg])
        
        symmetry = compute_bilateral_symmetry(
            left_max,
            right_max,
            metric_name=f"max_omega_{pair_name}"
        )
        bilateral_symmetry[f"max_omega_{pair_name}"] = symmetry

# Linear acceleration symmetry (max acceleration)
for pair_name, (left_seg, right_seg) in BILATERAL_PAIRS.items():
    if left_seg in lin_acc_mag and right_seg in lin_acc_mag:
        left_max = np.max(lin_acc_mag[left_seg])
        right_max = np.max(lin_acc_mag[right_seg])
        
        symmetry = compute_bilateral_symmetry(
            left_max,
            right_max,
            metric_name=f"max_lin_acc_{pair_name}"
        )
        bilateral_symmetry[f"max_lin_acc_{pair_name}"] = symmetry

print(f"\n✓ Bilateral symmetry computed for {len(bilateral_symmetry)} metrics")
print("\nSymmetry indices (1.0 = perfect, 0.0 = asymmetric):")
for key, data in sorted(bilateral_symmetry.items(), key=lambda x: x[1]["symmetry_index"]):
    print(f"  {key:30s}: {data['symmetry_index']:.3f} (L={data['left_value']:.1f}, R={data['right_value']:.1f}, diff={data['percent_diff']:.1f}%)")
```
